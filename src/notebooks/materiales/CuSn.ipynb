```python
#Entries are the basic unit for thermodynamic and other analyses in pymatgen.
#This gets all entries belonging to the Cu-Sn system.
entries = a.get_entries_in_chemsys(['Cu', 'Sn'])

#With entries, you can do many sophisticated analyses, like creating phase diagrams.
pd = pdmaker.PhaseDiagram(entries)
plotter = plotter.PDPlotter(pd)
plotter.show()
```

```python
# Import the basic libraries

# ASE system
import ase
from ase import Atom, Atoms
from ase import io
from ase.lattice.spacegroup import crystal

# Spacegroup/symmetry library
from pyspglib import spglib

# The qe-util package
#from qeutil import QuantumEspresso

# iPython utility function
from IPython.core.display import Image

```

## FINDSYM [link](http://stokes.byu.edu/iso/findsym.php)

```python
a=4.3596                                         # Lattice constant in Angstrom
cryst = crystal(['Cu', 'Sn'],                     # Atoms in the crystal
                [(0, 0.5, 0.5),  (0, 0, 0)], # Atomic positions (fractional coordinates)
                spacegroup=221,                  # International number of the spacegroup of the crystal
                cellpar=[a, a, a, 90, 90, 90])   # Unit cell (a, b, c, alpha, beta, gamma) in Angstrom, Degrees

```

```python
# Write the image to disk file
ase.io.write('images/crystal.png',       # The file where the picture get stored
             cryst,               # The object holding the crystal definition
             format='png',        # Format of the file
             show_unit_cell=2,    # Draw the unit cell boundaries
             rotation='115y,15x', # Rotate the scene by 115deg around Y axis and 15deg around X axis
             scale=30)            # Scale of the picture

# Display the image
Image(filename='images/crystal.png')
```

```python
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen import Lattice, Structure
import pymatgen.symmetry as sym
 
```

```python
coords = [[0, 0, 0], [0.0,0.5, 0.5], [0.5,0.5,0], [0.5,0,0.5]]
lattice = Lattice.from_parameters(a=4.3, b=4.3, c=4.3, alpha=90,
                                  beta=90, gamma=90)

    
struct = Structure(lattice, ["Sn", "Cu","Cu", "Cu"], coords)

s = sym.finder.Structure(lattice, ["Sn", "Cu","Cu", "Cu"], coords)

f = mg.symmetry.finder.SymmetryFinder(struct)

c = XRDCalculator()
c.show_xrd_plot(structure)
```

```python
print 'Space group:', spglib.get_spacegroup(cryst)
f.get_lattice_type(), f.get_crystal_system(), f.get_point_group(), f.get_spacegroup_number(), f.get_spacegroup_symbol(), f.get_hall()
```
