```python
import scipy             as scp
import numpy             as np
import control           as ctrl
import matplotlib.pyplot as mpl
```

```python
#  Parametros de la simulacion
DT = 0.001; Ttotal = 10; N = int(Ttotal/DT)

# Modelo de la planta
P = ctrl.tf([1],[1,2,0])
ctrl.bode(P);
```

```python
wgc = 10 #  # Desidered Bandwith
desideredPM = 60 

# Modelo del Controlador
Kp = 4; Kd = 2; Ki = .1; pd = 1
integ = ctrl.tf ([Ki],[1,0])
deriv = ctrl.tf ([Kd, 1], [0.1,1])
prop  = Kp
C = ctrl.parallel (deriv + prop, integ)

Ac,Bc,Cc,Dc = ctrl.matlab.ssdata(C)
print deriv, integ, C

ctrl.bode(C);
ctrl.bode(C*P);
```

```python
mag, phase, omega = ctrl.bode(C*P,[wgc],Plot=False)
print "mag: ", mag, "phase: ", phase, "omega: ", omega
```

```python
ph = phase[0]
if ph >= 0:
    ph = phase[0] -360
    
Phase = -180 + desideredPM
dPM = Phase - ph
print "Aditional phase from lead part", dPM
```

```python
#Lead Part
dPMrad = scp.pi*dPM/180
alfa = (1+scp.sin(dPMrad))/(1-scp.sin(dPMrad))
print "Alpha is: ", alfa  
```

```python
Tlead = 1/(scp.sqrt(alfa)*wgc)
Glead = ctrl.tf([alfa*Tlead,1],[Tlead,1])
print "Lead part is:", Glead
```

```python
#ctrl.bode(P)
ctrl.bode(C*Glead*P);
```

```python
mag, phase, omega = ctrl.bode(C*Glead*P,[wgc],Plot=False)
print "Phase at wgc is", phase[0]
K = 1/mag[0]
print "Gain to have MAG at wgc 0dB",K
print "Full controller", K*C*Glead
mag, phase, omega = ctrl.bode(K*C*Glead*P,[wgc],Plot=False)
print "omega: ", omega[0], "phase: ", phase[0], "mag: ", mag[0]
```

```python
ctrl.bode(K*C*Glead*P);
```

```python

#  Planta con realimentación unitaria PID 
l = ctrl.series (C,P)
sys = ctrl.feedback(l,1)

y,t = ctrl.matlab.step(sys, T = np.arange(0, Ttotal, DT) )
mpl.plot(t,y, label='C*Planta CL')
```

```python
#  Planta con realimentación unitaria controlador PD 
sys3 = ctrl.feedback((deriv + prop)*P,1)
y3,t = ctrl.matlab.step(sys3, T = np.arange(0, Ttotal, DT) )
mpl.plot(t,y3, label='(deriv+ prop)*Planta CL'); mpl.legend(loc='lower right')
mpl.show()
```
