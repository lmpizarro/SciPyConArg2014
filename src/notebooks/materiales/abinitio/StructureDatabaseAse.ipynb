```python
import ase
from ase.io import *

# iPython utility function
from IPython.core.display import Image

def createImageLattice (lattice):
    # Write the image to disk file
    ase.io.write('images/crystal.png',       # The file where the picture get stored
             lattice,               # The object holding the crystal definition
             format='png',        # Format of the file
             show_unit_cell=2,    # Draw the unit cell boundaries
             rotation='115y,15x', # Rotate the scene by 115deg around Y axis and 15deg around X axis
             scale=30)  
```

```python
#------------------------------------------------------------------
# Al bulk packed in the (111) direction (redefine lattice vectors)
from ase.lattice.cubic import FaceCenteredCubic
Al111 = FaceCenteredCubic(directions=[[1,-1,0], [1,1,-2], [1,1,1]],
            size=(1,1,1), 
            symbol='Al', 
            pbc=(1,1,1), 
            latticeconstant=4.05)

createImageLattice (Al111)
# Display the image
Image(filename='images/crystal.png')
```

```python
#------------------------------------------------------------------
# create a compound structure
from ase.lattice.compounds import Zincblende
GaN = Zincblende(directions=[[1,0,0], [0,1,0], [0,0,1]], 
            size=(1,1,1), 
            symbol=('Ga','N'), 
            latticeconstant=4.50)

          # Scale of the picture
```

```python
createImageLattice (GaN)
```

```python
Image(filename='images/crystal.png')
```

```python
#------------------------------------------------------------------
# complicate skutterudites structure with 32 atoms
from ase.lattice.spacegroup import crystal
skutterudite = crystal(('Co', 'Sb'), 
            basis=[(0.25,0.25,0.25), (0.0, 0.335, 0.158)], 
            spacegroup=204, 
            cellpar=[9.04, 9.04, 9.04, 90, 90, 90])
```

```python
createImageLattice (skutterudite)
Image(filename='images/crystal.png')
```

```python
#------------------------------------------------------------------
# create surface with vacuum
from ase.lattice.surface import *
slab = fcc111('Al', size=(2,2,3), a=4.05, vacuum=10.0)
createImageLattice (slab)
Image(filename='images/crystal.png')
```

```python
#------------------------------------------------------------------
# add an adsorbte on top
from ase.lattice.surface import *
slab = fcc111('Al', size=(2,2,3))
add_adsorbate(slab, 'H', 1.5, 'ontop')
slab.center(vacuum=10.0, axis=2)
createImageLattice (slab)
Image(filename='images/crystal.png')
```

```python
#------------------------------------------------------------------
# nanotube
from ase.structure import nanotube
cnt = nanotube(3, 3, length=6, bond=1.4, symbol='Si')
createImageLattice (cnt)
Image(filename='images/crystal.png')
```

```python
#------------------------------------------------------------------
# nanoribbons
from ase.structure import graphene_nanoribbon
gnr = graphene_nanoribbon(2, 6, type='zigzag', 
            saturated=True, C_H=1.1, C_C=1.4, 
            vacuum=6.0, 
            magnetic=True, 
            initial_mag=1.12, 
            sheet=False, 
            main_element='C', 
            saturate_element='H', 
            vacc=None)
createImageLattice (gnr)
Image(filename='images/crystal.png')
```

```python
write('Graphene.xyz', gnr, format='xyz')
write('Nanotube.xyz', cnt, format='xyz')
write('ZincBlende.xyz', GaN, format='xyz')
```

```python
from ase import *
from math import *
 
a = 3.189

atoms = Atoms([Atom('Sn', (0,       0,    0)),
               Atom('Cu', (0.5, 0.5, 0)),
               Atom('Cu',  (0,  0.5, 0.5)),
               Atom('Cu',  (0.5, 0, 0.5))])
 
cell = [(a,             0, 0),
        (0, a, 0),
        (0,             0, a)]
atoms.set_cell(cell, scale_atoms=True)
write('Atoms.cif', atoms, format='cif')
```

```python
createImageLattice (atoms)
Image(filename='images/crystal.png')
```

```python
# Spacegroup/symmetry library
from pyspglib import spglib

print 'Space group:', spglib.get_spacegroup(atoms)
```

```python
a = 4
Al111 = crystal(('Si', 'C'),
                       basis=[(0.0,0.0,0.0), (0.0, 0.5, 0.5)],
                       spacegroup=229, 
                       cellpar=[a, a, a, 90, 90, 90])

# Find a primitive cell of the structure
puc=spglib.find_primitive(Al111)
# Parameters of the unit cell
print 'Primitive unit cell (Angstrom, carthesian coordinates):\n', puc[0]
print '\nAtomic positions (fractional, crystalographic):\n', puc[1]
```

```python
# Create the primitive cell crystal
cryst_prim=Atoms(                         # Create a generic atomic structure
                 cell=puc[0],             # Unit cell (Angstrom)
                 scaled_positions=puc[1], # Atomic positions (fractional)
                 numbers=puc[2],          # Atomic numbers
                 pbc=True)                # Use Periodic Boundary Conditions
```

```python
# Check if the symmetry is still the same
# ver FINDSYM
print spglib.get_spacegroup(cryst_prim)
```

```python
createImageLattice (cryst_prim)
Image(filename='images/crystal.png')
```
