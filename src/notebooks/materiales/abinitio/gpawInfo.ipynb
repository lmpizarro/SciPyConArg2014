```python
import numpy as np
from ase.lattice import bulk
from gpaw import GPAW, PW, FermiDirac, MethfesselPaxton

a0 = 4.04
al = bulk('Al', 'fcc', a=a0)
cell0 = al.cell
# Write the image to disk file
ase.io.write('crystal.png',       # The file where the picture get stored
             al,               # The object holding the crystal definition
             format='png',        # Format of the file
             show_unit_cell=2,    # Draw the unit cell boundaries
             rotation='115y,15x', # Rotate the scene by 115deg around Y axis and 15deg around X axis
             scale=30)            # Scale of the picture

# Display the image
Image(filename='crystal.png')
```

```python
ecut = 300
al.calc = GPAW(mode=PW(ecut),
                   xc='PBE',
                   kpts=(8, 8, 8),
                   parallel={'band': 1},
                   basis='dzp')
eps = 0.01
al.cell = (1 + eps) * cell0
al.get_potential_energy()
```

```python
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline

xAlFcc = np.array([3.5, 3.6, 3.85, 3.95,  3.97,  3.98, 3.99, 4.00, 4.01, 4.04, 4.05, 4.25, 4.5])
yAlFcc = np.array([-13.262957, -14.739886, -16.544188, -16.719516, -16.727042, -16.727917, -16.726978, -16.724018, -16.719438, -16.695562, -16.684553, -16.193775, -15.144421])

for x,y in zip(xAlFcc, yAlFcc):
    print 'a: {0:.3f}  E: {1:.3f}.'.format(x,y)  

zAlFcc = np.polyfit(xAlFcc, yAlFcc, 4)

pAlFcc = np.poly1d(zAlFcc)

xp = np.linspace(3.45, 4.55, 100)
plt.plot(xAlFcc, yAlFcc, '.', xp, pAlFcc(xp), '-');
```

```python
deriv = np.gradient(pAlFcc(xp))
plt.plot(xp,deriv)
plt.grid()
coeff = np.polyfit(xp, deriv, 3)
raicesAlFcc = np.roots(coeff)
aMinAlFcc = 3.98099523
eMinAlFcc = pAlFcc (aMinAlFcc)
print 'AlFcc a: {0:.6f} E: {1:.6f}'.format(aMinAlFcc, eMinAlFcc)
raicesAlFcc
```

# Script para cálculo fcc 2 átomos 

```python
from ase import Atom, Atoms
from gpaw import GPAW
from gpaw.test import equal

from ase.lattice.spacegroup import crystal

a = 3.775

bulk = Atoms([Atom('Sn', (0, 0, 0)),
              Atom('Cu', (0, a*0.5, a*0.5)),
              Atom('Cu', (a*0.5, a*0.5, 0)),
              Atom('Cu', (a*0.5,0 ,a*0.5))    ],
              cell=(a, a, a),
              pbc=1)

n = 8
calc = GPAW(gpts=(2*n, 2*n, 2*n),
            nbands=1*32,
            kpts=(16, 16, 16),
            convergence={'eigenstates': 2.3e-2},
            xc='LDA')


bulk.set_calculator(calc)
e2 = bulk.get_potential_energy()
niter2 = calc.get_number_of_iterations()

print e2
```
