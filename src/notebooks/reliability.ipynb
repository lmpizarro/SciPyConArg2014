# Funciones de densidad de probabilidad (PFD)

## Caso discreto

### $P[X = x] = p(x) = p_{x}$

## Caso contínuo
### $ p[a \le x \le b] = \int_{a}^{b} {f(x)dx} $

```python
import scipy.stats as stats
from matplotlib import pyplot as plt
%matplotlib inline

media = 3.5
std   = 2.0
pdf = stats.norm(loc=media, scale=std)
x = np.arange(media-3*std,media+3*std,.1)
y = pdf.pdf(x)

plt.plot (x,y);
```

# Funciones de acumulación de distribución (CDF)

## $ F(x) = Pr[X \le x] = \alpha $

## Caso contínuo
### $ F(x) = \int_{-\infty}^{x} {f(\mu) d\mu} $

## Caso discreto

### $ F(x) = \sum_{i=0}^{x} {f(i)} $

```python
cdf = pdf.cdf(x)
plt.plot (x,cdf);
```

# Función inversa  de acumulación de distribución (PPF)

## $Pr[X \le G(\alpha)] = \alpha$
##### o de otra manera 
## $x = G(\alpha) = G(F(x)) $

```python
px = np.arange (0,1, .01)
ppf = pdf.ppf(px)
plt.plot (px,ppf);
```

# Función de sobrevida

### $S(x) = Pr[X > x] = 1 - F(x)$

```python
sf = pdf.sf(x)
plt.plot(x,sf);
```

# Función inversa de sobrevida

```python
isf = pdf.isf(px)
plt.plot(px,isf);
```

# Función de riesgo

### $h(x) = \frac {f(x)} {S(x)} = \frac {f(x)} {1 - F(x)}$

```python
h = pdf.pdf(x)/pdf.sf(x)
plt.plot (x,h);
```

# Función de riesgo acumulada

### $ H(x) = \int_{-\infty}^{x} {h(\mu) d\mu} $

### $H(x) = -\ln {(1 - F(x))}$

```python
H = -np.log (1 - pdf.cdf(x))
plt.plot(x,H);
```

# Exponencial

### \begin{array}{ll}
\mbox{PDF:}                 & f(t, \lambda) = \lambda e^{-\lambda t} \\
                            & \\
\mbox{CDF:}                 & F(t) = 1-e^{-\lambda t} \\
                            & \\
\mbox{Reliability:}         & R(t) = e^{-\lambda t} \\
                            & \\
\mbox{Failure Rate:}        & h(t) = \lambda  \\
                            & \\
\mbox{Mean:}                & \frac{1}{\lambda} \\
                            & \\
\mbox{Median:}              & \frac{\mbox{ln} 2}{\lambda} \cong \frac{0.693}{\lambda} \\
                            & \\
\mbox{Variance:}            & \frac{1}{\lambda^2} 
\end{array}

```python
expon1 = stats.expon(1)
expon2 = stats.expon(.01)
d1 = expon1.pdf(x)
d2 = expon2.pdf(x)
plt.plot(x,d1);
plt.plot(x,d2);
```

# Distribución de Weibull

#### \begin{array}{ll}
\mbox{PDF:}          & f(t, \gamma, \alpha) = \frac{\gamma}{t} \left( \frac{t}{\alpha} \right)^\gamma e^{- \left( \frac{t}{\alpha} \right)^\gamma} \\
                     & \\
\mbox{CDF:}          & F(t) = 1-e^{- \left( \frac{t}{\alpha} \right)^\gamma} \\
                     & \\
\mbox{Reliability:}  & R(t) = e^{- \left( \frac{t}{\alpha} \right)^\gamma} \\
                     & \\
\mbox{Failure Rate:} & h(t) =  \frac{\gamma}{\alpha} \left( \frac{t}{\alpha} \right) ^{\gamma-1} \\
                     & \\
\mbox{Mean:}         & \alpha \Gamma \left(1+\frac{1}{\gamma} \right) \\
                     & \\
\mbox{Median:}       & \alpha (\mbox{ln} \, 2)^{\frac{1}{\gamma}} \\
                     & \\
\mbox{Variance:}     & \alpha^2 \Gamma \left( 1+\frac{2}{\gamma} \right) - \left[ \alpha \Gamma \left( 1 + \frac{1}{\gamma}\right) \right]^2 
\end{array}

```python
def PDFWeill(t,alfa, gamma ):
    return (gamma/t)*np.power(t/alfa,gamma)*np.exp(-np.power(t/alfa,gamma))

vectPDFWeill = np.vectorize(PDFWeill)

gamma = 1.5; alfa = 8000
gamma2 = 1.5; alfa2 = 7000

t = np.arange (1, 3*alfa, 24)
fw =  vectPDFWeill (t, alfa, gamma)
fw2 =  vectPDFWeill (t, alfa2, gamma2)

plt.plot (t,fw2,)
plt.plot(t,fw);
```

```python
def CDFWeill (t, alfa, gamma):
    return 1 - np.exp(-np.power(t/alfa,gamma))

vectCDFWeill = np.vectorize(CDFWeill)

fwCDF = CDFWeill (t, alfa, gamma)
fwCDF2 = CDFWeill (t, alfa2, gamma2)
plt.plot (t, fwCDF2)
plt.plot(t,fwCDF);
```

```python
def FailRateWeill (t, alfa, gamma):
    return (gamma/alfa)*np.power(t/alfa,gamma-1)

vectFailRateWeill = np.vectorize (FailRateWeill)

fwFailureRate = vectFailRateWeill (t, alfa, gamma)
fwFailureRate2 = vectFailRateWeill (t, alfa2, gamma2)
plt.plot (t,fwFailureRate2);
plt.plot (t,fwFailureRate);
```

```python
def invCDFWeill(px,alfa, gamma ):
    return alfa*np.power(-np.log (1-px),1/gamma)

vectInvCDFWeill = np.vectorize(invCDFWeill)
tp = vectInvCDFWeill (px, alfa, gamma)
tp2 = vectInvCDFWeill (px, alfa2, gamma2)
plt.plot (px, tp2);
plt.plot (px, tp);
```

```python
N = 100000

r1 = np.random.random(N)
np.random.seed (4)
r2 = np.random.random(N)
tp1 = vectInvCDFWeill (r1, alfa, gamma)
tp2 = vectInvCDFWeill (r2, alfa2, gamma2)
exp = np.minimum (tp1, tp2) 
print (exp.mean(), exp.max(), exp.min(), exp.var(), exp.std())
```
