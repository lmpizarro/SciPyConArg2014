#### En los sistemas metálicos, la función que relaciona la energía con la distancia, obedece una relación universal parametrizable con 2 factores de escala

## $E(a) = E_c E^*(a^*)$

$E(a)$ es la energía total en función de la separación interatómica $a$

$E_c$ es la energía de interacción de equilibrio

$ E^*(a^*)$ es una función universal que describe la forma de la curva de interacción energética

##  $a^* = \frac {(a - a_m)} {l}$

La ordenada $a^*$, es una longitud afectada por un factor de escala

## $l =  \big (\frac {E_c}{12\pi B r_{ws}} \big) ^{1/2}$

$B$   es el bulk modulus

$r_{ws}$ es el radio de Wigner-Seitz de equilibrio

##  Aproximación para $r_{ws}$


## $r_{ws} = \big (\frac {3M}{4\pi \rho N_A} \big )^{1/3} $

$\rho$ es la densidad

$N_A$ es el número de avogadro

$M$, es la masa molar del elemento definida como la masa atómica del elemento $M_a$, multiplicada por $M_u = 1 \times 10^{-13} \frac{kg}{mol}$

# BFS

## Cambio de energía en la primer transformación, deformación

### $e_s^A = E_c^A[1 - (1 + a_s^*)exp(-a_s^*)]$ , (**)
donde
### $a_s^* = \frac{a-a_e^A}{l_A}$

 $a$: parámetro de red del cristal equivalente del átomo en cuestión
 
 $E_c^A$: es la energía de cohesión
 
 $a_e^A$:  el parámetro de de equilibrio
 $l_A$: factor de escala de longitud

```python
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division

%matplotlib inline
N = 100

rws  = 0  # equilibrium  value of Wigner-Seitz radius

aeA  = 4   # A equilibrium lattice constant 
lA   = 9   # scaling length
a    = np.array ([16.0*i/float(N) - 4.0  for i in range (N)])
asxx = (a-aeA)/lA
EcA  = 4
esA = 4*(1 -(1+asxx)*np.exp(-asxx))
plt.plot (asxx, esA);
```

## Cambio de energía en la segunda transformación, química

### $e_c^A = s_A E_c^A[1 - (1 + a_A^*)exp(-a_A^*)]exp(-a_A^*)$
donde
### $a_A^* = \frac{a_E-a_e^A}{l_A}$ 
### $ s_A = 1$ si $a_A^* > 0$, $ s_A = -1$ si $a_A^* < 0$,

$a_E$: se obtiene resolviendo ecuaciones trascendentales

```python
aE    = np.array ([16.0*i/float(N) - 4.0  for i in range (N)])
aAxx = (aE-aeA)/lA
sA  = np.copy(aAxx)
sA [sA < 0] = -1
sA [sA >= 0] = 1

ecA = sA*EcA*(1 - (1+aAxx)*np.exp(-aAxx))*np.exp(-aAxx) 
plt.plot (aAxx, ecA);
plt.plot (aAxx, ecA + esA, '.');
plt.plot (aAxx, esA, '*')
```

## Energía de deformación

### Determinación del parámetro de red equivalente $a$

### $NR_1^pexp(- \alpha R_1) + MR_2^pexp(- (\alpha  + \frac {1}{\lambda}) R_2) = \sum_i r_i^p exp [-(\alpha + S(r_i))r_i]$
donde
### $ S(r) = \frac {1}{2\lambda} (1 - cos([\frac {\pi (r-R_1^0)}{(R_2^0-R_1^0)}]))$


### $R_1^0 = c_1a_e$ y  $R_2^0 = c_2a_e$

$r_i$ es la distancia entre el $i^{th}$ vecino y el átomo en consideración y $N(M)$ es el número de vecinos cercanos y siguientes más cercanos, en el cristal perfecto. 

### $R_1 = c_1a$ y  $R_2 = c_2a$

donde $c_1$ y $c_2$ son tales que $R_1$ y $R_2$ son las distancias al vecino cercano y al siguiente más cercano.

Una vez determinado $a$, se calcula la energía de deformación con la ecuación (**)

# $\alpha$

Parámetro que refleja la estructura de la densidad electrónica en la zona de superposición, parámetro que se cálcula requiriendo acuerdo con la energía de formación de vacancias obtenida experimentalmente


```python
N = 10000
limit = 8
def lhseCu (R1):
    a = 3.6141
    N = 12 # cantidad de vecinos próximos
    M = 8  # cantidad de siguientes vecinos próximos
    dN  = 0.707; dM = 1;

    c2 = dM/dN

    pCu = 6; lambdaCu = 0.944; lCu = 0.272; alfaCu = 2.935;
    return N*np.power(R1,pCu)*np.exp(-alfaCu*R1) + M*np.power(c2*R1,pCu)*np.exp(- (alfaCu + 1/lambdaCu)*R1)
x = np.array ([limit*i/N for i in range (N)])
y = np.array([lhseCu(limit*i/N) for i in range (N)])
plt.plot (x,y);
```

# Energía Química

$\alpha_{AB} = \alpha_A + \Delta_{AB}$

$\alpha_A$ es el  valor del parámetro $\alpha$ para el metal puro A y $\Delta_{AB}$ es una corrección introducida por la presencia de un vecino de la especie B, $\Delta_{AB} = 0 $, si A = B.

### Valor experimental del calor de solución en el límite de dilución

$E^{AB}$: calor de solución de la impureza B en el host A

$E^{AB} = \frac {d}{dx} \Delta H _{x \to 1}$

$E^{BA} = \frac {d}{dx} \Delta H _{x \to 0}$



```python
# iPython utility function
from IPython.core.display import Image
import ase
from ase.io import *
def createImageLattice (lattice):
    # Write the image to disk file
    ase.io.write('crystal.png',       # The file where the picture get stored
             lattice,               # The object holding the crystal definition
             format='png',        # Format of the file
             show_unit_cell=2,    # Draw the unit cell boundaries
             rotation='115y,15x', # Rotate the scene by 115deg around Y axis and 15deg around X axis
             scale=30)  
```
