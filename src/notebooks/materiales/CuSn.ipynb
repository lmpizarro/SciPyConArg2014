```python
import pymatgen as mg


import pymatgen.matproj.rest as rest
import pymatgen.phasediagram.pdmaker as pdmaker
import pymatgen.phasediagram.plotter as plotter


%matplotlib inline


#This initializes the REST adaptor. You may need to put your own API key in as an arg.
a = rest.MPRester("xnID73SlfeROKikC")

cu = mg.Element("Cu")
print("Atomic mass of Cu is {}".format(cu.atomic_mass))
print("Cu has a melting point of {}".format(cu.melting_point))
print("Ionic radii for Cu: {}".format(cu.ionic_radii))

print  

sn = mg.Element("Sn")
print("Atomic mass of Sn is {}".format(sn.atomic_mass))
print("Sn has a melting point of {}".format(sn.melting_point))
print("Ionic radii for Sn: {}".format(sn.ionic_radii))


```

```python
#Entries are the basic unit for thermodynamic and other analyses in pymatgen.
#This gets all entries belonging to the Ca-C-O system.
entries = a.get_entries_in_chemsys(['Cu', 'Sn'])

#With entries, you can do many sophisticated analyses, like creating phase diagrams.
pd = pdmaker.PhaseDiagram(entries)
plotter = plotter.PDPlotter(pd)
plotter.show()
```

```python


# Import the basic libraries

# ASE system
import ase
from ase import Atom, Atoms
from ase import io
from ase.lattice.spacegroup import crystal

# Spacegroup/symmetry library
from pyspglib import spglib

# The qe-util package
from qeutil import QuantumEspresso

# iPython utility function
from IPython.core.display import Image

```

```python
a=4.3596                                         # Lattice constant in Angstrom
cryst = crystal(['Si', 'C'],                     # Atoms in the crystal
                [(0, 0, 0), (0.25, 0.25, 0.25)], # Atomic positions (fractional coordinates)
                spacegroup=216,                  # International number of the spacegroup of the crystal
                cellpar=[a, a, a, 90, 90, 90])   # Unit cell (a, b, c, alpha, beta, gamma) in Angstrom, Degrees
```

```python
# Write the image to disk file
ase.io.write('images/crystal.png',       # The file where the picture get stored
             cryst,               # The object holding the crystal definition
             format='png',        # Format of the file
             show_unit_cell=2,    # Draw the unit cell boundaries
             rotation='115y,15x', # Rotate the scene by 115deg around Y axis and 15deg around X axis
             scale=30)            # Scale of the picture

# Display the image
Image(filename='images/crystal.png')
```
