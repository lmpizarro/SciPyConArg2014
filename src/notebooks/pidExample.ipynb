```python
import control as ctrl
import numpy as np
import scipy as scp
#from matplotlib.pyplot import * # Grab MATLAB plotting functions
import matplotlib.pyplot as plt
```

```python
g = ctrl.tf ([1],[1,6,5])
```

```python
print g
```

```python
ctrl.bode (g);
```

```python
wgc = 10 #  # Desidered Bandwith
desideredPM = 60 
# PI part
Ti = 0.1
Gpi = ctrl.tf ([Ti, 1],[Ti, 0])
print "Parte PI:", Gpi
```

```python
ctrl.bode(g)
ctrl.bode(Gpi)
ctrl.bode(Gpi*g);
```

```python
mag, phase, omega = ctrl.bode(Gpi*g,[wgc],Plot=False)
```

```python
print "mag: ", mag, "phase: ", phase, "omega: ", omega
```

```python
ph = phase[0]
if ph >= 0:
    ph = phase[0] -360
    
Phase = -180 + desideredPM
dPM = Phase - ph
print "Aditional phase from lead part", dPM
```

```python
#Lead Part
dPMrad = scp.pi*dPM/180
alfa = (1+scp.sin(dPMrad))/(1-scp.sin(dPMrad))
print "Alpha is: ", alfa        
```

```python
Tlead = 1/(scp.sqrt(alfa)*wgc)
Glead = ctrl.tf([alfa*Tlead,1],[Tlead,1])
```

```python
print "Lead part is:", Glead
```

```python
ctrl.bode(g)
ctrl.bode(Gpi*Glead*g);
```

```python
mag, phase, omega = ctrl.bode(Gpi*Glead*g,[wgc],Plot=False)
print "Phase at wgc is", phase[0]
K = 1/mag[0]
print "Gain to have MAG at wgc 0dB",K
```

```python
ctrl.bode(K*Gpi*Glead*g);
```

```python
print "Full controller", K*Gpi*Glead
```

```python
mag, phase, omega = ctrl.bode(K*Gpi*Glead*g,[wgc],Plot=False)
print "omega: ", omega[0], "phase: ", phase[0], "mag: ", mag[0]
```

```python
gt = ctrl.feedback(K*Gpi*Glead*g,1)
Td = np.linspace(0,10,100)
```

```python
y,t = ctrl.step(gt)
plt.plot(t,y)
plt.grid()
```
