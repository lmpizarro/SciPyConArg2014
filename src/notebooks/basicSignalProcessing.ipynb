# Procesamiento de señales

Basado en los [tutoriales](http://docs.scipy.org/doc/scipy-dev/reference/tutorial/signal.html) de __scipy.signal__

```python
fs = 1000; N = 1000; amp = 2*np.sqrt(2); freq = 17; noise_power = 0.0001 * fs / 2
t = np.arange(N)/N
```

```python
# Define la señal
y = amp*np.sin(2*np.pi*freq*t) + 0.5*amp*np.sin(3*2*np.pi*freq*t) +  + 0.25*amp*np.sin(2*2*np.pi*freq*t)
r = np.random.normal(loc=0, scale=np.sqrt(noise_power), size=t.shape)
ymax = np.max(y)
ymin = np.min(y)
 
```

```python
mpl.plot (t,y, label='signal');
mpl.ylim(ymin*1.1, ymax*1.1);
mpl.xlim(0, 0.2);
mpl.legend(loc='lower right');
mpl.grid();
```

```python
yr = y + r

mpl.plot (t,yr, label='signal + ruido');
mpl.ylim(ymin*1.1, ymax*1.1);
mpl.xlim(0, 0.2);
mpl.legend(loc='lower right');
mpl.grid();
```

## Contenido espectral de la señal

```python
import scipy.signal as signal

f, Pwelch_spec = signal.welch(yr, fs, scaling='spectrum')
```

```python
mpl.semilogy(f, Pwelch_spec)
mpl.xlabel('frequency [Hz]')
mpl.ylabel('PSD')
mpl.grid()
mpl.show()
```

# Filtros FIR y Convolución

Filtro pasabajo 10 taps, frecuencia de corte en 100 Hz

```python
b1 = signal.firwin(10, 100, nyq=fs)
```

```python
z=np.convolve(y,b1)
```

```python
mpl.xlim(0,200)
mpl.plot(y)
mpl.plot(z);
```
