```python
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib.pyplot as mpl
from __future__ import division

%matplotlib inline
```

http://oneau.wordpress.com/2011/02/28/simple-statistics-with-scipy/

```python
e1 = stats.norm(loc=8700, scale=100.0)
sp = 4
e1lims=np.array([e1.mean()-sp*e1.std(),e1.mean()+sp*e1.std()])
x=np.arange(e1lims[0], e1lims[1],10)
y=e1.pdf(x)
mpl.plot(x,y);
```

```python
y=e1.cdf(x)
mpl.plot(x,y);
```

```python
x = np.arange(0, 1, 1/100)
y=e1.ppf(x)
mpl.plot(x,y);
```

```python
e2 = stats.norm(loc=8500, scale=20.0)
```

```python
N = 10000
exp = np.zeros(N)
for i in range(0,N):
   series = np.array([e2.ppf(np.random.random()),e1.ppf(np.random.random())])
   exp[i]=series.min()
```

```python
n, min_max, mean, var, skew, kurt = stats.describe(exp)
std = np.sqrt(var)
```

```python
n, min_max, mean, var, skew, kurt, std
```

```python
e3 = stats.norm(loc=mean, scale=std)
```

```python
x=np.arange(e1lims[0], e1lims[1],10)
y=e1.pdf(x)
mpl.plot(x,y);
y=e2.pdf(x)
mpl.plot(x,y);
y=e3.pdf(x)
mpl.plot(x,y);
```

```python
print "Serie"
print ("i1 std: %d mean: %d" % (e1.std(), e1.mean()))
print ("i2 std: %d mean: %d" % (e2.std(), e2.mean()))
print ("sc std: %d mean: %d" % (np.round((e3.std() + .5)), np.round(e3.mean() + .5)))
```
