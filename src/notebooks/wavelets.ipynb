```python
import numpy as np
import pywt
import matplotlib.pyplot as plt

%matplotlib inline
dir(pywt)
```

```python
N = 1000
f = 1
loc = 0
scale = .1
time = np.linspace (0, 1, N)
noise = np.array(np.random.normal(loc,scale,N))
data = np.sin(2*np.pi*f*time)
signal = noise + data
```

```python
plt.plot(time, signal);
```

```python
cA1, cD1 = pywt.dwt(signal, 'db2')
```

```python
plt.plot (cD1)
plt.plot (cA1)
```

```python
cD1 = np.zeros (cD1.shape)
reconst = pywt.idwt(cA1, cD1, 'db2')
```

```python
plt.plot (reconst)
```
