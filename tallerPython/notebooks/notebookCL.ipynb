# Cálculo del Controlador PID

```python
# Modelo del Controlador
Kp = 4; Kd = 2; Ki = .1; pd = 1
integ = ctrl.tf ([Ki],[1,0])
deriv = ctrl.tf ([Kd, 1], [0.1,1])
C = ctrl.parallel (deriv + Kp, integ)
```

# Planta con realimentación unitaria controlador P

```python
i = Image(filename='plantaProporcional.png')
i
```

```python
%%latex
La función transferencia es:
    
\begin{aligned}
H(s) = \frac {Kp \times P(s)} {(1+ Kp \times P(s))} 
\end{aligned}
```

```python
#  Planta con realimentación unitaria controlador P
sys2 = ctrl.feedback(Kp*P,1)
y,t = mtlb.step(sys2, T = np.arange(0, Ttotal, DT) )
mpl.plot(t,y, label='prop*Planta CL' ); mpl.legend(loc='lower right');
```

# Respuesta de la planta con controlador P + D

```python
%%latex
La función transferencia del controlador PID es:
    
\begin{aligned}
PID(s) = Kp + Ki \times \frac1s + Kd \times s
\end{aligned}

La función transferencia para la planta con el lazo de control cerrado es:

\begin{aligned}
H(s) = \frac { P(s) \times PID(s)} {(1+ P(s) \times PID(s))}

\end{aligned}
```

```python
#  Planta con realimentación unitaria controlador PD 
sys3 = ctrl.feedback((deriv + Kp)*P,1)
y3,t = mtlb.step(sys3, T = np.arange(0, Ttotal, DT) )
mpl.plot(t,y3, label='(deriv+ prop)*Planta CL'); mpl.legend(loc='lower right')
mpl.grid()
```

# Respuesta de la planta con controlador P + I + D

```python
#  Planta con realimentación unitaria PID 
l = ctrl.series (C,P)
sys = ctrl.feedback(l,1)

y,t = mtlb.step(sys, T = np.arange(0, Ttotal, DT) )
mpl.plot(t,y, label='C*Planta CL')
mpl.grid()
```

```python
wgc = 10 #  # Desidered Bandwith
desideredPM = 60 

Ac,Bc,Cc,Dc = mtlb.ssdata(C)
print deriv, integ, C

ctrl.bode(C);
ctrl.bode(C*P);
```

```python
mag, phase, omega = ctrl.bode(C*P,[wgc],Plot=False)
print "mag: ", mag, "phase: ", phase, "omega: ", omega
```

```python
ph = phase[0]
if ph >= 0:
    ph = phase[0] -360
    
Phase = -180 + desideredPM
dPM = Phase - ph
print "Aditional phase from lead part", dPM
```

```python
#Lead Part
dPMrad = scp.pi*dPM/180
alfa = (1+scp.sin(dPMrad))/(1-scp.sin(dPMrad))
print "Alpha is: ", alfa  
```

```python
Tlead = 1/(scp.sqrt(alfa)*wgc)
Glead = ctrl.tf([alfa*Tlead,1],[Tlead,1])
print "Lead part is:", Glead
```

```python
ctrl.bode(C*Glead*P);
```

```python
mag, phase, omega = ctrl.bode(C*Glead*P,[wgc],Plot=False)
print "Phase at wgc is", phase[0]
K = 1/mag[0]
print "Gain to have MAG at wgc 0dB",K
print "Full controller", K*C*Glead
mag, phase, omega = ctrl.bode(K*C*Glead*P,[wgc],Plot=False)
print "omega: ", omega[0], "phase: ", phase[0], "mag: ", mag[0]
```

```python
ctrl.bode(K*C*Glead*P);
```

# Respuesta de la planta con controlador P + I + D + Lead

```python
i = Image(filename='plantaLead.png')
i
```

```python
%%latex
La función transferencia del controlador en adelanto es:
    
\begin{aligned}
G_{lead}(s) = \frac {\alpha \times T_{lead} \times s + 1}{T_{lead} \times s +1}
\end{aligned}

La función transferencia para la planta con el lazo de control cerrado es:

\begin{aligned}
H(s) = \frac { P(s) \times PID(s) \times G_{lead}(s)} {(1+ P(s) \times PID(s) \times G_{lead}(s)}

\end{aligned}
```

```python
gt = ctrl.feedback(1*K*C*Glead*P,1)
y,t = mtlb.step(gt, T = np.arange(0, Ttotal, DT) )
mpl.plot(t,y)
mpl.grid()
```
