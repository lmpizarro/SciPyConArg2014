```python
import scipy as sc
import numpy as np
from __future__ import division
import matplotlib.pyplot as plt

%matplotlib inline

from IPython.display import display
from IPython.display import HTML
```

```python
# this is a wrapper that take a filename and publish an html <audio> tag to listen to it

import StringIO
import base64
import sys
import struct

def wavPlayer(data, rate):
    """ will display html 5 player for compatible browser

    The browser need to know how to play wav through html5.

    there is no autoplay to prevent file playing when the browser opens
    
    Adapted from SciPy.io.
    """
    
    buffer = StringIO.StringIO()
    buffer.write(b'RIFF')
    buffer.write(b'\x00\x00\x00\x00')
    buffer.write(b'WAVE')

    buffer.write(b'fmt ')
    if data.ndim == 1:
        noc = 1
    else:
        noc = data.shape[1]
    bits = data.dtype.itemsize * 8
    sbytes = rate*(bits // 8)*noc
    ba = noc * (bits // 8)
    buffer.write(struct.pack('<ihHIIHH', 16, 1, noc, rate, sbytes, ba, bits))

    # data chunk
    buffer.write(b'data')
    buffer.write(struct.pack('<i', data.nbytes))

    if data.dtype.byteorder == '>' or (data.dtype.byteorder == '=' and sys.byteorder == 'big'):
        data = data.byteswap()

    buffer.write(data.tostring())
#    return buffer.getvalue()
    # Determine file size and place it in correct
    #  position at start of the file.
    size = buffer.tell()
    buffer.seek(4)
    buffer.write(struct.pack('<i', size-8))
    
    val = buffer.getvalue()
    
    src = """
    <head>
    <meta http-equiv="Content-Type" content="text/html; charset=utf-8">
    <title>Simple Test</title>
    </head>
    
    <body>
    <audio controls="controls" style="width:600px" >
      <source controls src="data:audio/wav;base64,{base64}" type="audio/wav" />
      Your browser does not support the audio element.
    </audio>
    </body>
    """.format(base64=base64.encodestring(val))
    display(HTML(src))
```

```python
## some consstant for our audio file 

rate = 44100 #44.1 khz
fn = rate / 2
duration =np.power(2.0,17)/rate # in sec

# this will give us sin with the righ amplitude to use with wav files
normedsin = lambda f,t,a : a*2**14*np.sin(2*np.pi*f*t)

time = np.linspace(0,duration, num=int(rate*duration))
```

```python
# define A as a 440 Hz sin function 
la2    = lambda t : normedsin(2*440,t,1.0)
la1    = lambda t : normedsin(440,t,0.5)
la3    = lambda t : normedsin(450,t,0.5)
la4    = lambda t : normedsin(430,t,0.5)

la = la2(time) + la1 (time) + la3 (time) + la4 (time)

# look at it on the first 25 ms
plt.plot(time[0:1000], la[0:1000])

ampl = la.astype(np.int16)

# write the file on disk, and show in in a Html 5 audio player
wavPlayer(ampl, rate)
```

```python
fft_output = np.fft.fft(la)
Df = 2*fn/fft_output.shape[0]
freq = np.array( [i*Df for i in range(0,fft_output.shape[0])])
imag= fft_output.imag*fft_output.imag
real = fft_output.imag*fft_output.imag
mod = np.sqrt(real+imag)
```

```python
plt.plot(freq[0:4000] ,mod[0:4000]  );
```

# A 5 Hz waveform sampled at 500 Hz for 1 second

```python
from scipy import signal
 
t = np.linspace(0, 1, 500, endpoint=False)
plt.plot(t, signal.square(2 * np.pi * 5 * t))
plt.ylim(-2, 2);
```

# A pulse-width modulated sine wave:

```python
plt.figure()
sig = np.sin(2 * np.pi * t)
pwm = signal.square(2 * np.pi * 30 * t, duty=(sig + 1)/2)
plt.subplot(2, 1, 1)
plt.plot(t, sig)
plt.subplot(2, 1, 2)
plt.plot(t, pwm)
plt.ylim(-1.5, 1.5);
```

```python
t = np.linspace(0, 1, rate)
saw = 20000*signal.sawtooth(2 * np.pi * 440 * t,0.5)
squ = 20000*signal.square(2 * np.pi * 440 * t)
squ2 = 20000*signal.square(2 * np.pi * 440 * t,.2)
# plt.plot(t[0:1000], squ[0:1000])

signl = 1.0*saw + 1.0*squ + 0.2*squ2
plt.plot(t[0:1000], signl[0:1000]);
```

```python
ampl = signl.astype(np.int16)

#  show in in a Html 5 audio player
wavPlayer(ampl, rate)
```

```python
fft_output = np.fft.fft(ampl)
Df = 2*fn/fft_output.shape[0]
freq = np.array( [i*Df for i in range(0,fft_output.shape[0])])
imag= fft_output.imag*fft_output.imag
real = fft_output.imag*fft_output.imag
mod = np.sqrt(real+imag)
plt.plot(freq[0:22100]  ,mod[0:22100]   );
```

```python
def ADSR (sa, sd, val,ss, sr):
    at = [i/sa for i in range (sa)]
    de = [1 - val*i/sd for i in range (sd)]
    su = [val for i in range (ss)]
    re = [val - val*i/sr for i in range (sr)]
    return np.array(at+ de + su +re )

sig = np.sin(2 * np.pi * 12 * t) + np.sin(np.pi * 220 * t)
pwm = 20000*signal.square(2 * np.pi * 430 * t, duty=(sig + 1)/2)
mod = ADSR(4000,4000,0.5,6100,30000)
pwm=mod*pwm
```

```python
plt.plot(t, pwm);
```

```python
ampl = pwm.astype(np.int16)

#  show in in a Html 5 audio player
wavPlayer(ampl, rate)
```

```python
fft_output = np.fft.fft(ampl)
Df = 2*fn/fft_output.shape[0]
freq = np.array( [i*Df for i in range(0,fft_output.shape[0])])
imag= fft_output.imag*fft_output.imag
real = fft_output.imag*fft_output.imag
mod = np.sqrt(real+imag)
plt.plot(freq[0:22100]  ,mod[0:22100]   );
```
